## Motivations ( or previous work )
we are tackeling the problem of segmenting lesions and tissues in BUS from as a metric labeling problem using a discrete optimization approach. Here are some of the characteristics:

- Minimization technique : Graph-Cuts (GC)
- Data term : supervised Machine Learning (ML)
- Pair-wise term: {Markov/Conditional} Random fields (MRF/CRF)

### Current method PROS & CONS

1. Advantages 

    1. GC is fast.
    2. supervised ML offers a clean and powerful manner of designing the data term.

2. Disadvantages

    1. Only basic constrains over the final segmentation contour
    2. On ML task:
    
        1. Medical Image Analysis (MIA) is a big data problem (images are big)
        2. MIA databases have limited #images and limited Ground Truth (GT)
        3. Noisy features / unk. models highly overlapped
        4. GT noise, Labeling disagreement between experts
        
## Random thoughts

1. Using ML and Pattern Recognition (PR) to generate reliable stochastic models of the images elements from a reduced dataset (2.B.B) requires robust features and robust training policies. Which is not applicable (2.B.C, 2.B.D)

2. New ML approaches like sparse-coding and deep-learning are robust to noise in data, training, etc.. in expenses of more elaborated training techniques usually by minimizing a loss-function. These loss-fuctions can take many shapes and not all of them are appropiated for all the applications. However most of the time loss-function limitations are compensated by larger training sets and policies.

3. At the end of the day, when applying this new ML techniques, a metric that splits the data as much as possible is learned. 

4. Learning this metric from a small pool of noisy data with little knowledge about the features-lablels noise or their relation choosing the appropriated tool is crutial. 
.. - There's a relation between the features noise and the labeling noise (inconsistancy in experts delineations). Experts produce different delineations based on their criteria (intra-observer variation) but also their image interpretation (features interpretation --> inter-observer variation).

```python
# Define some class to facilitate working
class mySyntheticRealData:
    # Class constructors
    def __init__(self):
        self = mySyntheticRealData(modelMuX=0, modelMuY=0, modelSigmaX=1, modelSigmaY=1, modelSigmaXY=0, numSamples=100)
        
    def __init__(self,modelMuX, modelMuY, modelSigmaX, modelSigmaY, modelSigmaXY, numSamples):
        self.modelMuX = modelMuX
        self.modelMuY = modelMuY
        self.modelSigmaX = modelSigmaX
        self.modelSigmaY = modelSigmaY
        self.modelSigmaXY = modelSigmaXY
        self.data = np.random.multivariate_normal([modelMuX, modelMuY], [[modelSigmaX, modelSigmaXY], [modelSigmaXY, modelSigmaY]], numSamples)
        
    # Class displaying functions
    def __str__(self):
        return "Real-Data simulation: \n\tmean x:{0}, y:{1} \n\tstd x:{2}, y:{3}, xy:{4} \n\tnum data {5}\n".format( self.modelMuX, self.modelMuY,
                                                                                                                    self.modelSigmaX,self.modelSigmaY,self.modelSigmaXY,
                                                                                                                    self.data.shape[0])
    def show(self,targetAxis,scatterColor='k'):
        isoLinesColor = scatterColor
        # adjust figure
        axisCurrentValue = targetAxis.axis()
        axisNewValue = [ min( axisCurrentValue[0], self.data[:,0].min() ),
                         max( axisCurrentValue[0], self.data[:,0].max() ),
                         min( axisCurrentValue[1], self.data[:,1].min() ),
                         max( axisCurrentValue[1], self.data[:,1].max() )]
        targetAxis.axis (axisNewValue)
        # scatter Data plot
        x, y = self.data.T
        targetAxis.scatter(x,y, color=scatterColor, alpha=0.7)
        # iso-levels Data plot
        delta = (axisNewValue[1]-axisNewValue[0])/100
        X, Y = np.meshgrid( np.arange( axisNewValue[0], axisNewValue[1], delta ),
                            np.arange( axisNewValue[2], axisNewValue[3], delta ))
        Z = mlab.bivariate_normal( X, Y, self.modelSigmaX, self.modelSigmaY, self.modelMuX, self.modelMuY, self.modelSigmaXY )
        CS = targetAxis.contour(X, Y, Z/Z.max(), [0.8, 0.6, 0.36, 0.2], colors=isoLinesColor)
        #plt.clabel(CS,fmt[0]="xxxx",fmt[1]="yyyy")# ["20" "60" "90" "99"]) 
        
```

### Some functions

```python
def getWholeDataAxis (a,b):
    x = np.concatenate((a.data, b.data),axis=0)
    xLimits = [ x[:,0].min(), x[:,0].max(), x[:,1].min(), x[:,1].max() ]
    xExtra = min(xLimits[1]-xLimits[0], xLimits[3]-xLimits[2])/20                                       
    return xLimits + xExtra*np.array([-1, 1, -1, 1])

def drawPCAaxis(baseChange,targetAxis):
    base = [[-1,0],[1,0],[0,-1],[0,1]]
    newBase = baseChange.transform(6*np.array(base))
    x,y = newBase.T
    targetAxis.plot(x[0:2],y[0:2],'k-', linewidth=2)
    targetAxis.plot(x[2:4],y[2:4],'k-', linewidth=2)

def plotProjectedDataPDF(targetAxis,classBlueProjected,classRedProjected):
    colorScheme = ['b', 'r']
    peaks = [0, 0]
    f_of_D_min = np.concatenate((classBlueProjected, classRedProjected))[:,0].min()
    f_of_D_max = np.concatenate((classBlueProjected, classRedProjected))[:,0].max()
    
    Rm_subspaceLimits = [ f_of_D_min+0.1*(f_of_D_max-f_of_D_min),
                          f_of_D_max+0.1*(f_of_D_max-f_of_D_min)]
    X_plot = np.linspace( Rm_subspaceLimits[0], Rm_subspaceLimits[1], 1000)[:, np.newaxis]
    
    for i, currentClass in enumerate([classBlueProjected, classRedProjected]):
        X = (currentClass[:,0])[:,np.newaxis]
        kde = KernelDensity(kernel='gaussian', bandwidth=0.3).fit(X)
        log_dens = kde.score_samples(X_plot)
        (m,jj) = max((v,jj) for jj,v in enumerate(np.exp(log_dens)))
        #print "P(f(d))_max:{0} pos:{1} f(d):{2}".format(m, jj, X_plot[jj,0])
        peaks[i] = X_plot[jj,0]
        targetAxis.fill( X_plot[:, 0] ,
                         np.exp(log_dens) ,
                         colorScheme[i], alpha=0.2,
                       )
    currentAxisLimits = targetAxis.axis()
    targetAxis.stem(peaks,np.array(targetAxis.axis())[3]*np.array([1,1]),'k', markerfmt=' ')
    targetAxis.axis(currentAxisLimits)
    return peaks

def drawSamplingProbability(targetAxis,modelModes,sigma):
    colorScheme = ['b:', 'r:']
    currentAxisLimits = np.array(targetAxis.axis())
    displayScalingFactor = ((2*np.pi*1)**0.5)*currentAxisLimits[3]*.4
    #print ((2*np.pi*np.array(sigma))**-0.5)
    X_plot = np.linspace( currentAxisLimits[0], currentAxisLimits[1], 100)[:, np.newaxis]
    for currentSigma in sigma:
        for i, mu in enumerate(modelModes):
            Y_plot = ((2*np.pi*currentSigma)**-0.5)*np.exp(-((X_plot-mu)**2)*(2*currentSigma)**-1)
            targetAxis.plot(X_plot,displayScalingFactor*Y_plot,colorScheme[i], alpha=0.8)
            targetAxis.plot(X_plot,displayScalingFactor*Y_plot,'k:', alpha=0.8)

def sampleData(mu,std,data,nSamples=40,ProbLabelFlip=0):
    # data should be 1D
    selectedSamples = np.zeros((0,), dtype=np.integer)
    xCandidate = np.random.normal(mu,std,nSamples)[:,np.newaxis]
    for x in xCandidate:
        sampleDistance = (data.astype(float)-x)**2        
        for sampleIdx in selectedSamples:
            sampleDistance[sampleIdx] = np.inf
        (m,jj) = min((v,jj) for jj,v in enumerate(sampleDistance))
        selectedSamples = np.append(selectedSamples,jj) 
    flipLabel = np.random.uniform(0.0,1.0,nSamples)>(1-ProbLabelFlip)
    return selectedSamples,flipLabel   

def plotSampledData_inRm(targetAxis,data,sigmaId,pos,style):
    currentAxisLimits = np.array(targetAxis.axis())
    yi = data[:,1] - min(data[:,1])
    yi = (0.05*(yi/max(yi)))-0.2*sigmaId+pos
    currentAxisLimits[2]= min(yi)-0.1
    targetAxis.plot(data[:,0],yi,style)
    targetAxis.axis(currentAxisLimits)    
```

## Proposal 

1. Generate synthetic datasets from real data by sampling real data according to some criteria. Trying to simulate a progressive dificultie in the data.

    I think that what we are proposing has a relation with the monte-carlo-metropolis experiment. Plus varying the conditions to increase dificultie for each dataset.
    1. The inital data $D$ is in the hyperspace $R^N$ and each data point belongs to a single class $c \in C$ and is labeled as label $l \in L$ associated. Such that $\exists ! d | \<d,l_1\> & \<d,l_1\> & l_1 \ne l_2$.
    2. The metric is wanted to do something (representing the data) while keeping the data separated. Therefore, the metric is inteded to split the data on a lower dimensional space since the relation of feature dimension and data sparcity is quadratic. The ideal case is keep the data separated as much as possible with as little dimensionality as possible. 
    3. A linear mapping $f \in \{F \cap linear}$ where $F:R^N \to R^m$ is done. $f$ is intended such that everything still is a mess and the MAP of a point in the lower dimension mean nothing.
    4. This lower dimension space is used to sample a dataset $S$ from the initial data $D$ such that $S$ is a subset of $D$, $S \subset D$. In order to determine $S$ the classes are sampled in the lower dimension space. This sampling could be methodical, monte-carlo or any other one but it is chosen to be a gaussian sampling centered on the mode of the class in the lower space. Let $x$ be a coordinate in the reduced space $x \in R^m$. Then the probility of chosing a sample $d$ from class $c$ such that $x=f(d)$ in order to represent this class $c$ for the subset $S$ can be expresed as $P(x,c) \sim N( mode(f(d)), \sigma)$. 
    5. Hipothesis: 
    Using the samples in $S$ and mapping them to $R^m$, The separhability of the data should be AT WORST, as good as the separhability between the classes sampling strategy.

2. For each dataset try to split the data using different methodologies and see which is more constant. 

    1. By pushing this WORST CASE further and further meaning increasing sigma so that the sampling strategy gets closer and closer to the methodical or monte-carlo sampling strategy and $x$ become equiproval for all classes only metrics properly encoding the noise of the data and the labels should be able to keep spliting the samples.

## Before using real data, lest use completely sinthetic data and lets see what can we get from there

```python
np.random.seed(15031984)

# Define the data
blueData = mySyntheticRealData(0, 1, 1.2, 1, 0.8, 100)
redData  = mySyntheticRealData(1, 0, 1.3, 0.7, 0.3, 100)

# Show the data
fig = plt.figure(figsize=(14,5))
ax1 = fig.add_subplot(121)
ax1.axis( getWholeDataAxis( blueData, redData ) )
blueData.show(ax1,scatterColor='b')
redData.show(ax1,scatterColor='r')

# Calculate the Data projection
changeBase = PCA(n_components=2).fit(np.concatenate((blueData.data, redData.data),axis=0))
drawPCAaxis(changeBase, ax1)

# Project the data
blueDataProjected = changeBase.transform(blueData.data)
redDataProjected  = changeBase.transform(redData.data)

# Crjjeate the sampling
ax2 = fig.add_subplot(122)
ax2.cla()
modelModes = plotProjectedDataPDF(ax2,blueDataProjected,redDataProjected)
samplingSgimas = [.5, 1, 3]
drawSamplingProbability(ax2,modelModes,samplingSgimas)

for currentElement in [(blueData, blueDataProjected, modelModes[0],  0.025, 'b.'), 
                       (redData,  redDataProjected,  modelModes[1], -0.025, 'r.')]:
    for i, sigma in enumerate(samplingSgimas):
        projectedData = currentElement[1]
        xx,yy = sampleData( mu  = currentElement[2],
                            std = sigma,
                            data=projectedData[:,0]
                          )
        plotSampledData_inRm(ax2,currentElement[0].data[xx,:],i+1,currentElement[3],currentElement[4])
```
